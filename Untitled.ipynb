```python
from lxml import etree
import pandas as pd

df = pd.read_csv('C:/Users/ryanw/Downloads/MOCK_DATA.csv')
```

```python
def reverse_names(x):
    fname = x.split()[0]
    lname = x.split()[1]
    return '{}, {}'.format(lname, fname)

df['person_1'] = df['person_1'].apply(lambda x: reverse_names(x))
df['person_2'] = df['person_2'].apply(lambda x: reverse_names(x))
df['person_3'] = df['person_3'].apply(lambda x: reverse_names(x))

```

```python
df
```

```python
fake_xml_li = []

for idx, row in df.iterrows():
    p1 = row['person_1']
    p2 = row['person_2']
    p3 = row['person_3']
    t = row['title']
    d = row['date']
    year = str(d).split('-')[-1]
    jl = row['jel_list']
    doi = row['doi']
    publisher = row['publisher']
    journal = row['journal']
    tmp_item = create_record(p1=p1, p2=p2, p3=p3, t=t, d=d, y=year, jl=jl, doi=doi, publisher=publisher, journal=journal)
    fake_xml_li.append(tmp_item)
```

```python
etree.fromstring(''.join(fake_xml_li))
```

```python
with open('app/static/dummydata_large.xml', 'w') as f:
    f.write(''.join(fake_xml_li))
```

```python
def create_record(p1, p2, p3, t, d, y, jl, doi, publisher, journal):
    text = '''
    <record>
      <collection><![CDATA[/p16798coll7]]></collection>
      <pointer><![CDATA[8833]]></pointer>
      <filetype><![CDATA[url]]></filetype>
      <parentobject><![CDATA[-1]]></parentobject>
      <creata><![CDATA[{person_1}; {person_2}; {person_3}]]></creata>
      <title><![CDATA[{title}]]></title>
      <date><![CDATA[2017]]></date>
      <dmcreated><![CDATA[{date}]]></dmcreated>
      <jel><![CDATA[{jel_list}]]></jel>
      <subjec><![CDATA[Economics, Macro Economic activity]]></subjec>
      <publis><![CDATA[U.S. Government Agency]]></publis>
      <relatig><![CDATA[{journal}]]></relatig>
      <identia><![CDATA[{person_1}, {person_2}, {person_3} ({year}). "{title}&#44;" {journal} {year}-010. {publisher}. https://doi.org/10.17016/{doi}]]></identia>
      <citata><![CDATA[{person_1}, {person_2}, {person_3} ({year}). "{title}," {journal} {year}-010. {publisher}.]]></citata>
      <realtia><![CDATA[]]></realtia>
      <relatib><![CDATA[]]></relatib>
      <volume><![CDATA[{year}-{doi}]]></volume>
      <issue><![CDATA[]]></issue>
      <identi><![CDATA[]]></identi>
      <m1><![CDATA[m1a1; m1a2]]></m1>
      <board><![CDATA[Agency Division Name]]></board>
      <quarte><![CDATA[{year}Q1]]></quarte>
      <public><![CDATA[https://example.com/path/to/fakepdf{doi}.pdf]]></public>
      <link><![CDATA[http://example.com/path/to/metadata/fakemetadata{doi}]]></link>
      <doi><![CDATA[10.17016/{doi}]]></doi>
      <find><![CDATA[8835.url]]></find>
    </record>
    '''.format(person_1=p1, person_2=p2, person_3=p3, title=t, date=d, year=y, jel_list=jl, doi=doi, journal=journal, publisher=publisher)
    return text
```
